### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
# suma accediendo a cada campo de lista
suma_tienda1 = tienda["Precio"].sum()
suma_tienda2 = tienda2["Precio"].sum()
suma_tienda3 = tienda3["Precio"].sum()
suma_tienda4 = tienda4["Precio"].sum()

# Ingresos Totales
data_ingresos = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso_Total': [suma_tienda1, suma_tienda2, suma_tienda3, suma_tienda4]
}

df_ingresos_total = pd.DataFrame(data_ingresos)

print("--- Resumen de Ingresos por Tienda ---")
print(df_ingresos_total.to_markdown(index=False, numalign="left", stralign="left"))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.bar(df_ingresos_total['Tienda'], df_ingresos_total['Ingreso_Total'])

plt.title('Comparaci√≥n de Ingresos Totales')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')

plt.show()
```

# 2. Ventas por categor√≠a

```python
ventas_tienda1 = tienda['Categor√≠a del Producto'].value_counts()
ventas_tienda2 = tienda2['Categor√≠a del Producto'].value_counts()
ventas_tienda3 = tienda3['Categor√≠a del Producto'].value_counts()
ventas_tienda4 = tienda4['Categor√≠a del Producto'].value_counts()
```

```python
print("--- üõí Ventas por Categor√≠a - Tienda 1 ---")
print(ventas_tienda1.to_markdown(numalign="left", stralign="left"))

print("\n--- üõí Ventas por Categor√≠a - Tienda 2 ---")
print(ventas_tienda2.to_markdown(numalign="left", stralign="left"))

print("\n--- üõí Ventas por Categor√≠a - Tienda 3 ---")
print(ventas_tienda3.to_markdown(numalign="left", stralign="left"))

print("\n--- üõí Ventas por Categor√≠a - Tienda 4 ---")
print(ventas_tienda4.to_markdown(numalign="left", stralign="left"))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Definir los datos para el gr√°fico
etiquetas = ventas_tienda1.index   # Los nombres de las categor√≠as ser√°n las etiquetas
tama√±os = ventas_tienda1.values    # Los conteos de ventas ser√°n los tama√±os de las rebanadas

# 2. Crear la figura para el gr√°fico
plt.figure(figsize=(8, 8))

# 3. Generar el gr√°fico de torta
# 'autopct' formatea el porcentaje dentro de cada rebanada
plt.pie(
    tama√±os,
    labels=etiquetas,
    autopct='%1.1f%%',
    startangle=90  # Comienza el primer sector desde la parte superior
)

# 4. Asegurar que el gr√°fico sea un c√≠rculo perfecto
plt.axis('equal')

# 5. A√±adir T√≠tulo
plt.title('Distribuci√≥n de Ventas por Categor√≠a - Tienda 1', fontsize=14)

# 6. Mostrar el gr√°fico
plt.show()
```

# 3. Calificaci√≥n promedio de la tienda


```python
Calificaci√≥n_tienda1 = tienda['Calificaci√≥n'].mean()
Calificaci√≥n_tienda2 = tienda2['Calificaci√≥n'].mean()
Calificaci√≥n_tienda3 = tienda3['Calificaci√≥n'].mean()
Calificaci√≥n_tienda4 = tienda4['Calificaci√≥n'].mean()
```

```python
print("--- Calificaci√≥n Promedio Tienda 1 ---")
print(Calificaci√≥n_tienda1)

print("\n--- Calificaci√≥n Promedio Tienda 2 ---")
print(Calificaci√≥n_tienda2)

print("\n--- Calificaci√≥n Promedio Tienda 3 ---")
print(Calificaci√≥n_tienda3)

print("\n--- Calificaci√≥n Promedio Tienda 4 ---")
print(Calificaci√≥n_tienda4)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df_calificaciones = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificaci√≥n_Promedio': [
        Calificaci√≥n_tienda1,
        Calificaci√≥n_tienda2,
        Calificaci√≥n_tienda3,
        Calificaci√≥n_tienda4
    ]
})

# 2. Ordenar para mejor visualizaci√≥n (opcional)
df_calificaciones_ordenado = df_calificaciones.sort_values(
    by='Calificaci√≥n_Promedio',
    ascending=True
)

# --- GENERACI√ìN DEL GR√ÅFICO DE BARRAS HORIZONTAL ---
tiendas = df_calificaciones_ordenado['Tienda']
calificaciones = df_calificaciones_ordenado['Calificaci√≥n_Promedio']

# 4. Crear el gr√°fico
plt.figure(figsize=(10, 6))

# plt.barh() crea barras horizontales
plt.barh(tiendas, calificaciones, color='cornflowerblue')

# 5. A√±adir etiquetas de datos
for index, value in enumerate(calificaciones):
    # plt.text(valor_en_X, valor_en_Y, texto, alineaci√≥n_vertical)
    plt.text(value, index, f' {value:.2f}', va='center', fontsize=10)

# 6. T√≠tulos y Etiquetas
plt.title('‚≠êÔ∏è Comparaci√≥n de Calificaciones Promedio por Tienda', fontsize=16)
plt.xlabel('Calificaci√≥n Promedio (Escala 1 a 5)', fontsize=12)
plt.ylabel('Tienda', fontsize=12)

# 7. Establecer el rango del eje de calificaciones (X)
plt.xlim(0, 5)
plt.grid(axis='x', linestyle='--', alpha=0.6)

# 8. Mostrar el gr√°fico
plt.show()
```

# 5. Env√≠o promedio por tienda

```python
costo_envio_tienda1 = tienda['Costo de env√≠o'].mean()
costo_envio_tienda2 = tienda2['Costo de env√≠o'].mean()
costo_envio_tienda3 = tienda3['Costo de env√≠o'].mean()
costo_envio_tienda4 = tienda4['Costo de env√≠o'].mean()
```

```python
print("--- Env√≠o Promedio Tienda 1 ---")
print(costo_envio_tienda1)

print("\n--- Env√≠o Promedio Tienda 2 ---")
print(costo_envio_tienda2)

print("\n--- Env√≠o Promedio Tienda 3 ---")
print(costo_envio_tienda3)

print("\n--- Env√≠o Promedio Tienda 4 ---")
print(costo_envio_tienda4)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
-

df_costos_envio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo_Envio_Promedio': [
        costo_envio_tienda1,
        costo_envio_tienda2,
        costo_envio_tienda3,
        costo_envio_tienda4
    ]
})

# --- GENERACI√ìN DEL GR√ÅFICO DE L√çNEAS (COLUMNA/L√çNEA) ---

tiendas = df_costos_envio['Tienda']
costos = df_costos_envio['Costo_Envio_Promedio']


plt.figure(figsize=(10, 6))

plt.plot(tiendas, costos, marker='o', linestyle='-', color='darkorange')


plt.title('Costo de Env√≠o Promedio por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Costo de Env√≠o Promedio', fontsize=12)

from matplotlib.ticker import FuncFormatter
def formato_miles(x, pos):
    return '{:,.0f}'.format(x)
formato = FuncFormatter(formato_miles)
plt.gca().yaxis.set_major_formatter(formato)


for i, cost in enumerate(costos):
    plt.text(tiendas[i], cost + 300, f'{cost:,.0f}', ha='center')

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
